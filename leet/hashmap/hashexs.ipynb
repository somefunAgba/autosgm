```python

from collections import defaultdict

def find_longest_substring(s:str, k:int):
  counts = defaultdict(int)
  left = 0
  ans = 0
  for right in range(len(s)):
    counts[s[right]] += 1
    while len(counts) > k:
      counts[s[left]] -= 1
      if counts[s[left]] == 0:
        del counts[s[left]]
      left += 1
    
    ans = max(ans, right-left+1)
    
  return ans

s = "eceba"
find_longest_substring(s,2)
```

```python
from collections import defaultdict
from typing import List

def intersection(nums:List[List[int]]):
  # use an hashmap to count freq of occurence
  counts = defaultdict(int)
  for arr in nums:
    for x in arr:
      counts[x] +=1
    
  # see which appears n times
  n = len(nums)
  ans = []
  for key in counts:
    if counts[key] == n:
      ans.append(key)
      
  return ans

nums = [[3,1,2,4,5],[1,2,3,4],[3,4,5,6]]
intersection(nums)
```

```python
def equaloccur(s:str):
  counts = defaultdict(int)
  for c in s:
    counts[c] += 1
  freqs = counts.values()
  ans = len(set(freqs))==1
  
  return ans
```

```python
from collections import defaultdict

def grpAnagrams(strs:List[str])->List[List[str]]:
  groups = defaultdict(list)
  for s in strs:
    key="".join(sorted(s))
    groups[key].append(s)
    
  return groups.values()
```

```python
def mincardpick(cards:List[int]) -> int:
  acard= defaultdict(list)
  for idx,val in enumerate(cards):
    acard[val].append(idx)
  
  ans= float('inf')
  for key in acard:
    arr = acard[key]
    for i in range(len(arr)-1):
      right = arr[i+1]
      left = arr[i]
      ans = min(ans, right-left+1)
      
  if ans < float('inf'): pass
  else: ans = -1
  
  return ans
    
```

```python
def mincardpick(cards:List[int]) -> int:
  acard= defaultdict(int)
  ans= float('inf')
  for i in range(len(cards)):
    
    if cards[i] in acard:
      right = i
      left = acard[cards[i]]
      ans = min(ans, right-left+1)
      
    acard[cards[i]] = i
      
  if ans < float('inf'): pass
  else: ans = -1
  
  return ans
```

```python
def maxsum(nums:List[int]) -> int:
  
  def get_sum(num):
    sum = 0
    while num:
      sum += num % 10
      num //= 10
    return sum
  
  dic = defaultdict(int)
  ans = -1
  for num in nums:
    dig_sum = get_sum(num)
    if dig_sum in dic:
      ans = max(ans, num+dic[dig_sum])
    dic[dig_sum] = max(dic[dig_sum],num)
    
    return ans
```

```python
def eqrcpairs(grid:List[List[int]])-> int:
  def totuple(arr):
    return tuple(arr)
  
  rdic = defaultdict(int)
  for row in grid:
    rdic[totuple(row)] += 1
  
  cdic = defaultdict(int)
  for col in range(len(grid[0])):
    curr_col = []
    for row in range(len(grid)):
      curr_col.append(grid[row][col])
    cdic[totuple(curr_col)]+=1
  
  ans = 0
  for arr in rdic:
    ans += rdic[arr]*cdic[arr]
    
```
